```python
def weird_request(seq):
    for item in seq:
        seq.reverse()
    return(seq[1::])
```

```python
weird_request(["a","b","c"])
```

```python
assert weird_request(["a","b","c"]) == ["b","a"]
```

```python
weird_request([1,2,3])
```

```python
assert weird_request([1,2,3]) == [2,1]
```

```python
weird_request([1])
```

```python
assert weird_request([1]) == []
```

```python
weird_request([])
```

```python
assert weird_request([]) == []
```
